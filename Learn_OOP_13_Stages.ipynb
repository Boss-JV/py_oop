## ‚úÖ Stage 4
**What you learn:** Add __init__ method to automate attribute setup

```python
# üëâ Code for Stage 4 goes here
class Student:
    def __init__(self, name, age, courses,pocket_money,location):
        self.name = name
        self.age = age
        self.courses = courses
        self.pocket_money = pocket_money
        self.location = location

def display_student_info(student):
    print("Name:", student.name)
    print("Age:", student.age)
    print("Courses:", ", ".join(student.courses))       
    print("Pocket Money",student.pocket_money)
    print("Location:",student.location)
    print("-" * 20)
    
    
student1 = Student("Alice", 20, ["Math", "English"],100,"Bangkok")
student2 = Student("Bob", 22, ["Physics", "History"],500,"Nonthaburi")

display_student_info(student1)
display_student_info(student2)
# Feel free to modify or expand it
```
